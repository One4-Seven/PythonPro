### 数据解析 Beautiful
- 基本步骤：
  - 指定URL
  - 发送请求
  - 获取响应数据
  - 解析数据（BS4表达式）
  - 存储数据库

### Beautiful在爬虫中的使用流程
- 创建Beautiful对象
  - 本地：Beautiful（'open（'本地的html文件'）', 'lxml'）
  - 网络：Beautiful（'网络请求到的页面数据', 'lxml'）
- 核心思想
  - 将HTML文档转换成Beautiful对象
  - 调用该对象中的属性和方法进行HTML文档指定内容的定位查找。

```python
from bs4 import BeautifulSoup

fp = open('Save_files/Test.html', encoding = 'UTF-8')
soup = BeautifulSoup(fp, 'lxml')

soup.div  # 根据标签名进行查找 #

soup.a['href']  # 获取标签属性值 #

soup.p.string  # 获取内容 #

soup.find('div', class_="tang")  # 获取符合条件的标签 #
soup.findAll('a', limit=2)

soup.select('div > img')[0]  # select函数的层级选择器 #
soup.select('div > ul > li')
soup.select('div  li')
```

### 以小说网为例爬取三国演义的详情
#### 具体实现步骤
- 获取整个网页的源码(HTML)
- 转换为BS4的处理对象并获取章节标题和对应内容URL
- 通过新的URL结合BS4处理并爬取章节内容
- 保存本地数据

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.shicimingju.com/book/sanguoyanyi.html'
headers = {  # 设置请求头 #
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36 Edg/128.0.0.0'
}
def get_content(url):  # 根据地址爬取章节内容 #
    content_page = requests.get(url = url, headers = headers).content
    soup = BeautifulSoup(content_page, 'lxml')
    div = soup.find('div', class_ = 'chapter_content')
    return div.text

page_text = requests.get(url = url, headers = headers).content
soup = BeautifulSoup(page_text, 'lxml')

a_list = soup.select('.book-mulu > ul > li > a')  # a_list列表中存储的是一系列的a标签对象 #
fp = open('Save_files/Romance_of_the_three_kingdoms.txt', 'w', encoding = 'UTF-8')
for a in a_list:
    title = a.string  # 获取章节标题 #
    content_url = 'https://www.shicimingju.com/' + a['href']  # 拼接内容地址 #
    content = get_content(content_url)
    fp.write(title + ':' + content + '\n\n\n')  # 本地存储 #

fp.close()
print('Save Success')
```
